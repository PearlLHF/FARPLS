```python
def custom_distance(vector1, vector2):
    vector1 = np.array(vector1, dtype=float)
    vector2 = np.array(vector2, dtype=float)
    
    # Euclidean distance
    euclidean_dist = np.linalg.norm(vector1 - vector2)

    # Manhattan distance (sum of absolute differences)
    manhattan_dist = np.sum(np.abs(vector1 - vector2))
    
    # combine distances with equal weights
    distance = 0.5 * euclidean_dist + 0.5 * manhattan_dist
    
    return distance

# create a new column 'dist' for each pair: pre-processed
def calculate_distance_for_pair(df):    
    df['dist'] = df.apply(lambda row: custom_distance(row['trajectory1_feature'], row['trajectory2_feature']), axis=1)
    
```
