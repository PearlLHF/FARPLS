```python
import pandas as pd
import matplotlib.pyplot as plt
# from matplotlib.lines import Line2
import os
# 从CSV文件中读取数据
baseline_data = pd.read_csv("../questionnaire/baseline.csv")
experiment_data = pd.read_csv("../questionnaire/experiment.csv")

```

```python
colorset = { '#038C73','#89D9CA','#a3daff','#D9BA82','#dedcee','#A65F21','#fffcf0'}
def cum_size_until_4(data_counts):
    cum_size = 0
    for value, size in data_counts.items():
        try:
            numeric_value = int(value)
        except ValueError:
            continue
        if numeric_value <= 4:
            cum_size += size
        
    if 4 in data_counts:
        cum_size -= data_counts[4] / 2
    return cum_size
```

```python


def plot_stacked_bar_for_column(column_name):
    # 计算每个评分的频率
    baseline_counts = baseline_data[column_name].value_counts().sort_index()
    experiment_counts = experiment_data[column_name].value_counts().sort_index()

    # 创建堆叠条形图
    fig, ax = plt.subplots(figsize=(10, 3.5)) ##A65F21
    colorset_mapping = {1: '#E54E35', 2: '#FA9857', 3: '#FED884', 4: '#FEFFBE', 5: '#D1EC86', 6: '#91D068', 7: '#3FAA59'} 
   
    legend_handles = [plt.Rectangle((0,0),1,5, facecolor=colorset_mapping[i]   ) for i in range(1, 8)]
    legend_labels = [
        '1-Strongly disagree',
        '2-Disagree',
        '3-Neutral',
        '4-Agree',
        '5-Somewhat Agree',
        '6-Agree',
        '7-Strongly agree'
    ]
    # legend_handles = [Line2D([0], [0], color=gray_scale_mapping[i], lw=6, linestyle='-', marker='s', markersize=10, markerfacecolor=gray_scale_mapping[i], markeredgecolor='black') for i in range(1, 8)]
    # 绘制数据
    def plot_data(label, data_counts, position,offset = 0):
        cum_size = 0 + offset
         # 定义数字评分级别到文本的映射
        
        for  idx,  (value, size) in enumerate(data_counts.items()):
            ax.barh(label, size, left=cum_size, height= 0.4, color=colorset_mapping.get(value, '0.5'),   linewidth=2.2)
            text_x = cum_size + (size / 2)
            ax.text(text_x, position,str(value), ha='center', va='center', color='black', fontsize=10)
            cum_size += size
            
    cum_baseline = cum_size_until_4(baseline_counts)
    cum_experiment = cum_size_until_4(experiment_counts)
    diff = cum_baseline - cum_experiment
    if diff > 0:
        offset_experiment = diff
        offset_baseline = 0
    else:
        offset_baseline = -diff
        offset_experiment = 0
   

    plot_data('Experiment', experiment_counts, 0,offset_experiment)
    plot_data('Baseline', baseline_counts, 1,offset_baseline)
   
     # Add legend below the plot, using bbox_to_anchor to move it further down
    ax.legend(legend_handles, legend_labels, title='Rating Level', loc='upper center', 
              bbox_to_anchor=(0.5, 0), ncol=len(legend_labels), frameon=False)
    plt.subplots_adjust(bottom=0.3)
    # 在左侧添加标签
    ax.text(-2, 0, 'FARPLS', ha='right', va='center', fontsize=12, fontweight='bold')
    ax.text(-2, 1, 'Baseline', ha='right', va='center', fontsize=12, fontweight='bold')

    # 设置图形属性
    ax.set_xlabel('Number of Users')
    ax.set_title(f'Horizontal Stacked Barplot of {column_name} for Baseline and Experiment Data')
    ax.axis('off')
    ax.axvline(cum_baseline + offset_baseline, ymin=0.5, ymax=1, linestyle='--')  # baseline
    ax.axvline(cum_experiment + offset_experiment, ymin=0, ymax=0.5, linestyle='--')  # experiment
    pdf_filename = f"./fig/Barplot2/{column_name}.pdf"
    fig.savefig(pdf_filename, bbox_inches='tight', format='pdf')
    # 保存图像
    fig.savefig(f"./fig/Barplot2/{column_name}.png", bbox_inches='tight')
    
    plt.show() 
    plt.close(fig)  # 关闭图形，防止显示大量图形

# 确保当前目录存在
if not os.path.exists('./plots'):
    os.makedirs('./plots')

# 遍历每一列并绘图
for column in baseline_data.columns:
    plot_stacked_bar_for_column(column)
   
```
