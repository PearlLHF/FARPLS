```python
def reset(uid):
    url = f'http://127.0.0.1:5000/reset/baseline/{uid}'
    response = requests.get(url)
    response
    

def query(uid):
    url = f'http://127.0.0.1:5000/query/baseline/{uid}'
    video_pair = requests.get(url)
    vid_pair =video_pair.json()
    try:
        vid_pair = [v.split("/")[-2] for v in vid_pair]
    except:
        return vid_pair
    return vid_pair

def submit(vid_pair, pref_vid, uid):
    url = f'http://127.0.0.1:5000/submit/baseline/{uid}'
    # pref_vid = vid_pair[1]
    response = requests.post(url, data={
        "video1": vid_pair[0], 
        "video2": vid_pair[1],
        "pref_vid": pref_vid
    })
    return response
```

```python
import random
from tqdm.notebook import tqdm
```

```python
reset("U05")
reset("U03")
```

```python
uid_list = ["U05", "U03"]
with tqdm(range(240), total=240) as pbar:
    for i in pbar:
        uid = random.choice(uid_list)
        vid_pair = query(uid)
        if isinstance(vid_pair, str):
            print(vid_pair)
            break
        pref_vid = random.choice(vid_pair + ["no pref"])
        pbar.set_postfix({"pref_vid": pref_vid})
        response = submit(vid_pair=vid_pair, pref_vid=pref_vid, uid=uid)
        if response.status_code != 200:
            print(i, response.text)
            break
        if response.text != f"database/baseline/{uid}_pref.csv":
            print(i, response.text)
            if response.text == '"No more video pairs to label."\n':
                uid_list.remove(uid)
                if len(uid_list) == 0:
                    break
```
